```python
dir_in = 'txt_cut'

char_range = []

for filename in sorted(os.listdir(dir_in)):
    with open(os.path.join(dir_in, filename), 'r') as book:
        size = len(book.read())
        if size < 10:
            print(filename)
        char_range.append(size)
char_range = pd.Series(char_range)
```

```python
import re
re.sub('[\n]+', '\n', '\n\n\n \n\n')
```

```python
def gen():
    for i, filename in enumerate(np.array(sorted(os.listdir('txt')))[(char_range < 100000)]):
        try:
            with open(os.path.join('txt_cut', filename), 'r') as book:
                b = book.read().replace('\n', ' ')
                print(filename, len(b), i)
                yield b[-1000:]
        except:
            continue
```
