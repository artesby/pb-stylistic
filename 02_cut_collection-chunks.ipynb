```python
import os
import re
```

```python
CHUNK_SIZE = 15000
MIN_CHUNK_SIZE = 10000
dir_in = 'txt'
dir_out = 'txt_cut_chunks' + str(int(CHUNK_SIZE/1000))
MIN_SIZE = 14000
MAX_SIZE = 60000


if not os.path.exists(dir_out):
    os.makedirs(dir_out)
```

```python
for filename in sorted(os.listdir(dir_in)):
    chunks = []
    with open(os.path.join(dir_in, filename), 'r') as book:
        content = book.read()
        print(filename, len(content))
        content = content.replace('?..', '?').replace('!..', '!').replace('...', '').replace('--', '')
        content = re.sub('[]+', '  ', content)
        content = re.sub('[_]+', '_', content)
        content = re.sub('[\n]+', ' \n ', content)
        content = re.sub('[*]+', '*', content)
        content = re.sub('\xa0', ' ', content)
        if len(content) < MIN_SIZE:
            continue
#         if len(content) > MAX_SIZE:
#             cur = MAX_SIZE
#             next_char = content[cur]
#             while next_char.isalpha() and len(content) > cur:
#                 cur += 1
#                 next_char = content[cur]
#             content = content[:cur]
        n_chunks = min(MAX_SIZE, len(content)) // CHUNK_SIZE
        rs = len(content) % CHUNK_SIZE
        cur = 0
        for i in range(n_chunks):
#             print(i, len(content))
            last = cur
            cur = CHUNK_SIZE * (i+1)
            next_char = content[cur]
            cur_char = content[cur-1]
            while cur_char.isalpha() and next_char.isalpha() and len(content) > cur:
                cur += 1
                next_char = content[cur]
            chunk = content[last:cur]
            chunks.append(chunk)
        if rs > MIN_CHUNK_SIZE and n_chunks < MAX_SIZE / CHUNK_SIZE:
            chunks.append(content[cur:])
        

    for i, ch in enumerate(chunks):
        print(len(ch))
        with open(os.path.join(dir_out, filename + f'_chunk{i}'), 'w') as book:
            book.write(ch)
        
```
